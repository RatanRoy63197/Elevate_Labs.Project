```python
import pandas as pd

df=pd.read_csv(r"C:\Users\Xervice5433\Downloads\car_prices.csv\car_prices.csv")

print(df.head())
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df['make'] = df['make'].fillna(
    df['make'].mode()[0]
)

df['model'] = df['model'].fillna(
    df['model'].mode()[0]
)

df['trim'] = df['trim'].fillna(
    df['trim'].mode()[0]
)

df['body'] = df['body'].fillna(
    df['body'].mode()[0]
)

df['transmission'] = df['transmission'].fillna(
    df['transmission'].mode()[0]
)

df['vin'] = df['vin'].fillna(
    df['vin'].mode()[0]
)

df['condition'] = df['condition'].fillna(
    df['condition'].mode()[0]
)

df['odometer'] = df['odometer'].fillna(
    df['odometer'].mode()[0]
)

df['color'] = df['color'].fillna(
    df['color'].mode()[0]
)

df['interior'] = df['interior'].fillna(
    df['interior'].mode()[0]
)

df['seller'] = df['seller'].fillna(
    df['seller'].mode()[0]
)

df['mmr'] = df['mmr'].fillna(
    df['mmr'].mode()[0]
)

df['sellingprice'] = df['sellingprice'].fillna(
    df['sellingprice'].mode()[0]
)

df['saledate'] = df['saledate'].fillna(
    df['saledate'].mode()[0]
)
```

```python
print(df.isnull().sum())
```

```python
print(df.head(6))
```

```python
df['saledate'] = pd.to_datetime(df['saledate'], dayfirst=True, errors='coerce')
df['saledate'] = df['saledate'].dt.strftime('%d/%m/%Y')
```
